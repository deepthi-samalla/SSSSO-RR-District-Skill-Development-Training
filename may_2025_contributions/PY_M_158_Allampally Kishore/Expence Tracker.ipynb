```python
import csv
from datetime import datetime

# File to store transactions
filename = 'expenses.csv'

# Ensure file exists with headers
try:
    with open(filename, 'x', newline='') as f:
        writer = csv.writer(f)
        writer.writerow(['Date', 'Amount', 'Category', 'Description'])
except FileExistsError:
    pass

# Function to add a transaction
def add_transaction():
    date = input("Enter date (YYYY-MM-DD): ")
    amount = float(input("Enter amount: "))
    category = input("Enter category (e.g. Food, Salary): ")
    description = input("Enter description: ")
    
    with open(filename, 'a', newline='') as f:
        writer = csv.writer(f)
        writer.writerow([date, amount, category, description])
    print("‚úÖ Transaction added.")

# Function to view summary
def view_summary():
    income = 0
    expense = 0
    with open(filename, 'r') as f:
        reader = csv.DictReader(f)
        for row in reader:
            amt = float(row['Amount'])
            if amt >= 0:
                income += amt
            else:
                expense += abs(amt)
    print(f"\nüìä Summary:")
    print(f"Total Income: ‚Çπ{income:.2f}")
    print(f"Total Expenses: ‚Çπ{expense:.2f}")
    print(f"Net Balance: ‚Çπ{income - expense:.2f}")

# Menu loop
while True:
    print("\nüíº Expense Tracker")
    print("1. Add Transaction")
    print("2. View Summary")
    print("3. Exit")
    choice = input("Choose an option (1‚Äì3): ")
    
    if choice == '1':
        add_transaction()
    elif choice == '2':
        view_summary()
    elif choice == '3':
        print("üëã Goodbye!")
        break
    else:
        print("‚ùå Invalid choice.")
```
