```python
# Example 1: Create connection and cursor
conn = sqlite3.connect(':memory:')
cursor = conn.cursor()
```

```python
# Example 2: Create table
cursor.execute("CREATE TABLE students (id INTEGER, name TEXT, score REAL)")
conn.commit()
```

```python
# Example 3: Resize image
resized = cv2.resize(img, (100, 100))
```

```python
plt.imshow(cv2.cvtColor(resized, cv2.COLOR_BGR2RGB))
```

```python
0-255
```

```python
# Example 4: Draw rectangle
rect = cv2.rectangle(resized, (0, 0), (40, 40), (0,78, 200), 2)
```

```python
plt.imshow(cv2.cvtColor(rect, cv2.COLOR_BGR2RGB))
```

```python
# Example 5: Save image
cv2.imwrite('output.jpg', rect)
```

```python
# Example 7: Gaussian blur
blurred = cv2.GaussianBlur(resized, (5, 5), 0)
plt.imshow(cv2.cvtColor(blurred, cv2.COLOR_BGR2RGB))
```

```python
# Example 8: Edge detection
edges = cv2.Canny(resized, 1000,100)
plt.imshow(cv2.cvtColor(edges, cv2.COLOR_BGR2RGB))
```

```python
# Example 9: Image thresholding
_, thresh = cv2.threshold(cv2.cvtColor(resized, cv2.COLOR_BGR2GRAY), 100, 255, cv2.THRESH_BINARY)
```

```python
plt.imshow(thresh)
```

```python
# Example 10: Contour detection
contours, _ = cv2.findContours(thresh, cv2.RETR_TREE, cv2.CHAIN_APPROX_SIMPLE)
img2 = cv2.drawContours(resized, contours, -1, (0, 255, 0), 2)
```

```python
plt.imshow(cv2.cvtColor(img2, cv2.COLOR_BGR2RGB))
```

```python
# Example 6: Capture from webcam
cap = cv2.VideoCapture(0)
ret, frame = cap.read()
cv2.imshow('Webcam', frame)
cap.release()
cv2.destroyAllWindows()
```

### Web Scrapping

```python
import requests
from bs4 import BeautifulSoup
```

```python
'https://meebhoomi.ap.gov.in/''
```

```python
# Example 1: Simple GET request
response = requests.get('https://www.ikea.com/in/en/')
```

```python
response.status_code
```

```python
# Example 2: Parse HTML content
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
soup.title.string
```

```python
# Example 3: Find all links
links = soup.find_all('a')

```

```python
for link in links:
    print(link.get('href'))
```

```python
# Example 4: Find element by id
header = soup.find_all(id="c5f809b0-2ee9-11f0-8589-b9660497574f")
# print(header.text if header else "No header found")
```

```python
len(header)
```

```python
header.text
```

```python
items = soup.find_all(class_="pub__btn__label")
print([item.text for item in items])
```

```python
[item.text for item in items]
```

```python
# Example 6: Scrape multiple pages (pagination example)
for page in range(1, 4):
    url = f"https://example.com/page/{page}"
    r = requests.get(url)
    sp = BeautifulSoup(r.text, 'html.parser')
    print(f"Page {page} title:", sp.title.string)
```

```python
# Example 1: Simple GET request
response = requests.get('https://www.scrapethissite.com/pages/forms/?page_num=1')
```

```python
response.status_code
```

```python
soup = BeautifulSoup(response.text, 'html.parser')
```

```python
# Example 7: Extract table data
table = soup.find('table')
rows = table.find_all('tr') if table else []
for row in rows:
    cols = [ele.text.strip() for ele in row.find_all('td')]
    print(cols)
```
